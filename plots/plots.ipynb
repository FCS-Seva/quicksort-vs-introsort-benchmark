```python
import pandas as pd
import matplotlib.pyplot as plt

csv_path = "sort_results.csv"

df = pd.read_csv(csv_path, sep=';')

df.head()
```

```python
types = ["random", "reversed", "almost"]

for t in types:
    df_sub = df[(df["type"] == t) & (df["algo"] == "quick")]

    plt.figure(figsize=(8, 5))
    plt.plot(df_sub["n"], df_sub["time_ms"])
    plt.xlabel("Размер массива n")
    plt.ylabel("Время, мс")
    plt.title(f"Quick sort: {t} массивы")
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
threshold = 16
types = ["random", "reversed", "almost"]

for t in types:
    df_sub = df[
        (df["type"] == t) &
        (df["algo"] == "intro") &
        (df["threshold"] == threshold)
    ]

    plt.figure(figsize=(8, 5))
    plt.plot(df_sub["n"], df_sub["time_ms"])
    plt.xlabel("Размер массива n")
    plt.ylabel("Время, мс")
    plt.title(f"Intro sort (threshold={threshold}): {t} массивы")
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
threshold = 16
types = ["random", "reversed", "almost"]

for t in types:
    df_merge = df[(df["type"] == t) & (df["algo"] == "quick")]
    df_hybrid = df[
        (df["type"] == t) &
        (df["algo"] == "intro") &
        (df["threshold"] == threshold)
    ]

    plt.figure(figsize=(8, 5))
    plt.plot(df_merge["n"], df_merge["time_ms"], label="Quick sort")
    plt.plot(df_hybrid["n"], df_hybrid["time_ms"], label=f"Intro Sort, threshold={threshold}")
    plt.xlabel("Размер массива n")
    plt.ylabel("Время, мс")
    plt.title(f"{t} массивы: quick vs intro (threshold={threshold})")
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()
```
